```python
from utils import loading_strumok_tables
from strumok import Strumok
```

```python
tables = loading_strumok_tables("strumok_tables.c")
cipher = Strumok(tables)

print("r1 =", cipher.r1)
print("r2 =", cipher.r2)
print("s =", cipher.s)
```

```python
key_512 = [
    0, 0, 0, 0,
    0, 0, 0, 0x8000000000000000,
]
iv_256 = [0, 0, 0, 0]

cipher.init_512(key_512, iv_256)

print("r1 =", hex(cipher.r1))
print("r2 =", hex(cipher.r2))
print("s[0] =", hex(cipher.s[0]))
print("s[15] =", hex(cipher.s[15]))
```

```python
z0 = cipher.get_number64()
print("z0 =", hex(z0))
print("new r1 =", hex(cipher.r1))
print("new r2 =", hex(cipher.r2))
print("new s[0] =", hex(cipher.s[0]))
print("new s[15] =", hex(cipher.s[15]))
```
